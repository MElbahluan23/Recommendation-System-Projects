```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Get The Data

```python
DATA_PATH="./Datasets/"
credits=pd.read_csv(os.path.join(DATA_PATH,"tmdb_5000_credits.csv"))
movies=pd.read_csv(os.path.join(DATA_PATH,"tmdb_5000_movies.csv"))
```

```python
credits.head()
```

```python
movies.head()
```

```python
print("Credits Shape: ",credits.shape)
print("Movies Shape: ",movies.shape)
```

```python
credits.rename(columns={"movie_id":"id"},inplace=True)
credits.columns
```

```python
movies_df_merge=movies.merge(credits,on='id',suffixes=('_movies', '_credits'))
print("Final Shape: ",movies_df_merge.shape)
movies_df_merge.head()
```

```python
movies_df_merge.columns
```

```python
movies_df_merge.drop(columns=['homepage','title_movies','title_credits','status','production_companies'],inplace=True)
```

```python
movies_df=movies_df_merge.copy()
movies_df.head()
```

```python
movies_df.info()
```

# Using Weighted average for each movie's Average Rating

```python
v=movies_df.vote_count
R=movies.vote_average
C=movies.vote_average.mean()
m=movies['vote_count'].quantile(0.70)
```

```python
m
```

```python
movies_df["weighted_avg"]=((R*v)+(C*m))/(v+m)
```

```python
movies_df.head()
```

```python
movies_df.columns
```

```python
movies_sorted_ranking=movies_df.sort_values("weighted_avg",ascending=False)
movies_sorted_ranking[['original_title','vote_average','vote_count','popularity','weighted_avg']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python

plt.figure(figsize=(15,6))
plt.xlim(5,10)
sns.barplot(x=movies_sorted_ranking.weighted_avg.head(10),y=movies_sorted_ranking.original_title.head(10))
plt.title("Best Movies by average votes",weight='bold')
plt.xlabel("Weighted avg",weight='bold')

plt.ylabel("Movies title",weight='bold')
plt.show()
```

```python
movies_sorted_ranking_popularity=movies_df.sort_values("popularity",ascending=False)
movies_sorted_ranking_popularity[['original_title','vote_average','vote_count','popularity','weighted_avg']]
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=movies_sorted_ranking_popularity.popularity.head(10),y=movies_sorted_ranking_popularity.original_title.head(10))
plt.title("Best Movies by popularity",weight='bold')

plt.xlabel("popularity",weight='bold')

plt.ylabel("Movies title",weight='bold')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
scaled_movies=pd.DataFrame(scaling.fit_transform(movies_df[['weighted_avg','popularity']]),columns=['weighted_avg','popularity'])
```

```python
# scoring top movies from popularity & weighted_avg same importance

movies_df['score_based_50%']= (scaled_movies*.5).sum(axis=1)
```

```python
Top20Movies=movies_df.sort_values("score_based_50%",ascending=False)[['original_title','weighted_avg','popularity','score_based_50%']].head(20)
```

```python
plt.figure(figsize=(15,6))
plt.xlim(.3,1)
plt.title("Top Recommended Movies to Watch",weight='bold')
sns.barplot(x='score_based_50%',y='original_title',data=Top20Movies[['score_based_50%','original_title']].head(10))
plt.xlabel("Score",weight='bold')

plt.ylabel("Movies title",weight='bold')
plt.show()
```
